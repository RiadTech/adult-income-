```python
# 17. Replace Salary Values With 0 and 1
df["income"].unique()[0]
```

```python
df["income"].unique()[1]
```

```python
df["income"].value_counts()
```

```python
df['encoded_income'] = df["income"].apply(lambda x: 0 if x == '<=50K' else 1)
```

```python
df.head()
```

```python
# 18. Which Workclass Getting The Highest Salary?
highest_sal = df.groupby('workclass')['encoded_income'].mean().sort_values(ascending=False)

print('highest salary ->', highest_sal.keys()[0], '->', highest_sal.values[0])
```

```python
# 19.How Has Better Chance To Get Salary greater than 50K Male or Female?
per_income = df.groupby('gender')['encoded_income'].mean().sort_values(ascending= False)
per_income
```

## Memory Optimization

```python
# 20. Covert workclass Columns Datatype To Category Datatype ( memory optimization)
df["workclass"] = df["workclass"].astype('category')
df.dtypes
```

```python
df.info() # from 5.6+ MB to 4.5+ MB 
print("Saved Memory: 1.1 MB")
```
